# model building

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, ExtraTreesRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score, GridSearchCV
```

```python
X_min = temp_min.drop('Next_Tmin', axis =1)
y_min = temp_min.Next_Tmin

X_max = temp_max.drop('Next_Tmax', axis =1)
y_max = temp_max.Next_Tmax
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_min = scaler.fit_transform(X_min)
X_max = scaler.fit_transform(X_max)
```

```python
Xmin_train, Xmin_test, ymin_train, ymin_test = train_test_split(X_min, y_min, test_size=0.33, random_state=42)
for i in [Xmin_train, Xmin_test, ymin_train, ymin_test]:
    print(i.shape)
```

```python
Xmax_train, Xmax_test, ymax_train, ymax_test = train_test_split(X_max, y_max, test_size=0.33, random_state=42)
for i in [Xmax_train, Xmax_test, ymax_train, ymax_test]:
    print(i.shape)
```

```python
lr = LinearRegression()
ls = Lasso()
rg = Ridge()
en = ElasticNet()
rfr = RandomForestRegressor()
abr = AdaBoostRegressor()
etr = ExtraTreesRegressor()
gbr = GradientBoostingRegressor()
xgbr = XGBRegressor()
knr = KNeighborsRegressor()
```

## models for minimum temperature

```python
models = [lr,ls,rg,en,rfr,abr,etr,gbr,xgbr,knr]
for i in models:
    i.fit(Xmin_train,ymin_train)
    pred_i = i.predict(Xmin_test)
    
    print(f'Model Report {i}')
    print('MAE:', mean_absolute_error(ymin_test,pred_i))
    print('MSE: ', mean_squared_error(ymin_test,pred_i))
    print('RMSE:', np.sqrt(mean_squared_error(ymin_test,pred_i)))
    print('R2 SCORE:', r2_score(ymin_test,pred_i))

    print('xxxxx'*20)
    print('\n')
```

OBS-: We selected   models for further evaluation
   1. linear regression ---> 0.83
   2. ridge reg ---> 0.83
   3. random forest regressor ---> 0.89
   4. extra tree regressor ----> 0.90

## models for maximum temperature

```python
models = [lr,ls,rg,en,rfr,abr,etr,gbr,xgbr,knr]
for i in models:
    i.fit(Xmax_train,ymax_train)
    pred_i = i.predict(Xmax_test)
    
    print(f'Model Report {i}')
    print('MAE:', mean_absolute_error(ymax_test,pred_i))
    print('MSE: ', mean_squared_error(ymax_test,pred_i))
    print('RMSE:', np.sqrt(mean_squared_error(ymax_test,pred_i)))
    print('R2 SCORE:', r2_score(ymax_test,pred_i))

    print('xxxxx'*20)
    print('\n')
```
